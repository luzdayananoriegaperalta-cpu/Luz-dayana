## ¿Qué es Python?

Python es un lenguaje de programación de **alto nivel**, lo que significa que su sintaxis es muy parecida al lenguaje humano (específicamente al inglés). Es famoso por ser:

* **Legible:** Es fácil de aprender y escribir.
* **Versátil:** Se usa desde inteligencia artificial hasta automatización de maquinaria.
* **Comunitario:** Tiene miles de "librerías" (herramientas listas para usar) creadas por expertos de todo el mundo.

## ¿Por qué es importante en la Ingeniería Agroindustrial?

En la agroindustria, ya no basta con saber de procesos químicos o mecánicos; el manejo de **datos** y la **automatización** son la clave de la eficiencia. Python es vital porque:

1. **Agricultura de Precisión:** Permite analizar datos de suelos, clima y sensores para decidir exactamente cuánto fertilizante o agua usar.
2. **Control de Calidad:** Puedes programar sistemas de visión artificial que detecten frutas dañadas en una banda transportadora.
3. **Optimización de Procesos:** Ayuda a calcular las rutas logísticas más cortas o a predecir cuándo una máquina va a fallar (mantenimiento predictivo).

## 3 Ejemplos de código útiles en Agroindustria

Aquí tienes tres aplicaciones prácticas y sencillas:

### 1. Cálculo de la Humedad en Base Seca

Para procesos de secado de granos, es fundamental conocer la humedad.

```
# Cálculo de contenido de humedad
peso_inicial = 500  # gramos
peso_seco = 420     # gramos

humedad_base_seca = ((peso_inicial - peso_seco) / peso_seco) * 100

print(f"La humedad en base seca es del {humedad_base_seca:.2f}%")
```

### 2. Clasificación Automática por Peso

Imagina un sensor en una línea de producción que clasifica aguacates por su tamaño.

```
peso_fruta = 250  # gramos

if peso_fruta < 150:
    categoria = "Pequeño (Calibre C)"
elif 150 <= peso_fruta <= 300:
    categoria = "Mediano (Calibre B)"
else:
    categoria = "Grande (Calibre A)"

print(f"El fruto pertenece a la categoría: {categoria}")
```

### 3. Estimación de Rendimiento de Cosecha

Un cálculo rápido para proyectar la producción por hectárea.

```
plantas_por_metro = 8
peso_promedio_fruto = 0.45  # kg
metros_cuadrados = 10000    # 1 Hectárea

rendimiento_total = plantas_por_metro * peso_promedio_fruto * metros_cuadrados

print(f"El rendimiento estimado de la cosecha es de {rendimiento_total} kg por hectárea.")
```
